# Converting the cancer dataset to a dataframe

```python
#df_cancer = pd.DataFrame(data = [cancer.data, cancer.target], columns = list(cancer.feature_names) + ['target'])
#Converting the sklearn.dataset cancer to a dataframe
cancer_df = pd.DataFrame(data=np.c_[cancer.data, cancer.target], columns=list(cancer.feature_names) + ['target'])
cancer_df.head()
```

```python
cancer_df.info()
```

# Showing the class distribution and also the df into X and y


```python
#This shows the class distribution, how many instances of malignant and how many of benign

df = cancer_df['target'].value_counts()
df.index = ['benign', 'malignant']
#Splitting the DataFrame into X(the data) and y(the labels)
X = cancer_df.iloc[:, 0:30]
y = cancer_df['target']
   
```

# Splitting into training set and test set

```python
#Splitting into training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

# Fitting Knn with Training Data, Predicting class label for test data, Getting the score of Knn Classifier using X_test and y_test

```python
#Using KNeighboursClassifier, fit a k-nearest neighbours(knn) classifier with X_train, y_train and using one nearest neighbour
#n_neighbours =1
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
#Using knn classifier, we predict the class label using the mean value for each feature
means = cancer_df.mean()[:-1].values.reshape(1, -1)
#Using knn classifier, predict the class labels for the test set X_test
knn.predict(means)
knn.predict(X_test)
#The score of Knn Classifier using X_test and y_test
knn.score(X_test, y_test)  
```
